```
 url ='http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json'

movie_df = pd.DataFrame()
for index in final_data['movieCd']:
  param = {'',   # 발급받은 키
          'movieCd' : '{}'.format(index)}
  rq = requests.get(url, params = param)
  j = rq.json()
  try:
    df = pd.DataFrame(columns = j['movieInfoResult']['movieInfo'].keys(),data = [j['movieInfoResult']['movieInfo'].values()])
    movie_df=pd.concat([movie_df,df], ignore_index=True)
  except:
    pass
```

```
del movie_df['movieNmEn']
del movie_df['movieNmOg']
del movie_df['prdtStatNm']
del movie_df['nations']
del movie_df['genres']
del movie_df['audits']
del movie_df['staffs']
```
