```
import os
import sys
import urllib.request
import pandas as pd
import json
import re
import requests

start = 1
end = 100
movie_df = pd.DataFrame()
for start_index in range(start,end):
  print('*')
  url ='http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json'
  param = {'key' : '',   # 발급받은 키
          'curPage' : '{}'.format(start_index),
          'openStartDt' : ['2020','2019','2018','2017','2016','2015'],
          'itemPerPage' : '100'}
  rq = requests.get(url, params = param)
  j = rq.json()
  for i in range(0,100):
    try:
      df = pd.DataFrame(columns = j['movieListResult']['movieList'][i].keys(),data = [j['movieListResult']['movieList'][i].values()])
      movie_df=pd.concat([movie_df,df], ignore_index=True)
    except:
      pass
movie_df
```
